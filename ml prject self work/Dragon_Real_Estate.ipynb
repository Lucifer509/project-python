Now we want our random in a fixed propotion example as we know 'CHAS' is only 0 or 1 we not want our random to select all the 1
value and not leave it for test data or vise versa. So,  we need our random test and train data in a fixed ratio of each feature
present. To do that we use:

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['CHAS']):         # we made ratio in according to 'CHAS'
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

## Creating pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler 
from sklearn.impute import SimpleImputer
     # now as scale of our feature are different which can be a problem we use StandardScaler to make all scale same.
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    #  add as many as you want in your pipeline
    ('std_scaler',StandardScaler())
])
```

```python
housing_prossed = my_pipeline.fit_transform(housing)
```

```python
housing_prossed.shape
```

## Now pipeline is created all we have to do is selecting the mode

```python
# from sklearn.linear_model import LinearRegression
# from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()


model.fit(housing_prossed, housing_labels)
```

Before we train our model on whole data we run it on small data

```python
some_features = housing.iloc[:5]
some_lables = housing_labels.iloc[:5]
```

```python
some_prossed_data = my_pipeline.fit_transform(some_features)  # prossed our small data through pipeline.
```

```python
model.predict(some_prossed_data)
```

## Error in model

```python
from sklearn.metrics import mean_squared_error
housing_predict = model.predict(housing_prossed)
mse = mean_squared_error(housing_labels,housing_predict)
rmse = np.sqrt(mse)
```

```python
rmse                
```

In case of decisionTreeRegressor our model get overfit.
Now to solve this problem we can change the way we give data to model.
instead of giving whole training data we divide it in 10 part then we only give 9 part in second turn we add removed part
but remove another part according to this way our model pesrform better.

## Cross validation - using better technique 

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model, housing_prossed, housing_labels, scoring='neg_mean_squared_error', cv =10) 
    # here cv is how many part you want to devide.
rsme_score = np.sqrt(-score)
```

```python
rsme_score    # give 10 value becaue we run 10 time so our model predict 10 time.
```

```python
def print_scores(scores):
    print("scores",scores)
    print("Mean", scores.mean())
    print("Standard_deviation", scores.std())
```

```python
print_scores(rsme_score)
```
